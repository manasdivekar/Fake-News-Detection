```python
df_fakenews = pd.read_csv("Fake.csv")
```

```python
# Merging the data sets of fake and correct news

df_merge = pd.concat([df_fakenews , df_correctnews], axis = 0)
df_merge.head(15)
```

```python
df_merge.columns
```

```python
# Removing those columns which are not required for detecting the fake news

data = df_merge.drop(['title','subject','date'] , axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample( frac = 1)
```

```python
data.head() 
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
# In this we try preprocess data and clean and optimize it by changing the entire data into lowercase,Remove square brackets and their contents,Remove non-word characters,remove url


def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '' , text)
    text = re.sub("\\W" , " " , text)
    text = re.sub("https?://\S+|www\.\S+" , '' , text)
    text = re.sub("<.*?>+" , '' , text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '' , text)
    text = re.sub("\n" , '' , text)
    text = re.sub("\w*\d\w" , '' , text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
# Training the dataset 

x_train , x_test ,y_train , y_test = train_test_split(x,y , test_size =0.30)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train , y_train)
```

```python
prediction_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test , y_test)
```

```python


print(classification_report(y_test , prediction_lr))
```

```python
# Decision tree classification

from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train , y_train)
```

```python
prediction_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test , y_test)
```

```python
print(classification_report(y_test , prediction_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train , y_train)
```

```python
prediction_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test , y_test)
```

```python
print(classification_report(y_test , prediction_rf))
```

```python
def output_label(n):
    if n == 0:
        return "FAKE NEWS!!"
    elif n==1:
        return "NOT A FAKE NEWS!!!!!"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    prediction_LR = LR.predict(new_xv_test)
    prediction_DT = DT.predict(new_xv_test)
    prediction_RF = RF.predict(new_xv_test)

    return print("\n\n LR Prediction: {} \n RF Prediction: {}\n".format(output_label(prediction_LR[0]),
                                                                                              
                                                                                              output_label(prediction_RF[0])))
```
